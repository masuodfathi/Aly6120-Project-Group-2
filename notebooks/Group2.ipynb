## what is “not sensible"?

Based on outputs, the following values are suspicious:

1. Negative market share 
2. Negative 5G subscribers
3. 5G coverage above 100%
4. 640 duplicate rows

***Everything else seems normal***

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
df.describe()
```

```python
#Create flags for bad values before fixing
df["Negative_MarketShare_Flag"] = df["Market Share (%)"] < 0
df["Negative_5GSubscribers_Flag"] = df["5G Subscribers (millions)"] < 0
df["Coverage_Above_100_Flag"] = df["Regional 5G Coverage (%)"] > 100
```

```python
#Fix negative and out-of-bounds values
df.loc[df["Market Share (%)"] < 0, "Market Share (%)"] = 0
df.loc[df["5G Subscribers (millions)"] < 0, "5G Subscribers (millions)"] = 0
df["Regional 5G Coverage (%)"] = df["Regional 5G Coverage (%)"].clip(0, 100)
```

```python
#Convert Quarter (e.g., "Q1") to month (Q1→01, Q2→04, Q3→07, Q4→10)
quarter_to_month = {"Q1": "01", "Q2": "04", "Q3": "07", "Q4": "10"}
df["Month"] = df["Quarter"].map(quarter_to_month)
df["Date"] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Month"] + "-01")

```

```python
#Create a Quarter_Index for each Product Model
#It let us track the number of quarters since a product model was launched
# We can know how long a product model has been on the market
# This can be useful for analyzing trends related to product lifecycle
# For example, we can see if market share tends to decline after a certain number of quarters
df = df.sort_values(by="Date")
df["Quarter_Index"] = df.groupby("Product Model").cumcount() + 1
```

```python
df["Is_5G"] = df["5G Capability"].apply(lambda x: 1 if x == 'Yes' else 0)
df.head(30)
df["5G_Rediness_Score"] = (((df["Avg 5G Speed (Mbps)"] / df["Avg 5G Speed (Mbps)"].max()) * 0.5 )+ 
                          (df["Regional 5G Coverage (%)"] / 100 * 0.5))
df.head(30)
```
